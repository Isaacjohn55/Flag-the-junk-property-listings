### Using Pipelines for processing data

```python
p1=pdPipeline([
    ('var_select',VarSelector(['PriceIndex9','PriceIndex8', 'PriceIndex7', 'PriceIndex4', 'PriceIndex1', 'PriceIndex6',
                               'InsurancePremiumIndex','PriceIndex3', 'PriceIndex5', 'NormalisedPopulation', 'PriceIndex2'])),
    ('object_to_numeric',convert_to_numeric()),
    ('missing_trt',DataFrameImputer())
])
p2=pdPipeline([
    ('var_select',VarSelector(['InteriorsStyle', 'Material', 'Agency', 'AreaIncomeType',
                               'EnvRating','ExpeditedListing','PRIMEUNIT', 'Channel', 'PlotType', 'Architecture',
                               'Region', 'SubModel', 'Facade', 'State', 'BuildYear', 'RegionType','Zip', 'PropertyAge'])),
    ('missing_trt',DataFrameImputer()),
    ('create_dummies',get_dummies_Pipe(500))
])
p3=pdPipeline([
    ('var_select',VarSelector(['ListDate'])),
    ('covert_to_datetime',convert_to_datetime()),
    ('cyclic_dates',cyclic_features())
])
p4=pdPipeline([
    ('var_select',VarSelector(['data']))
])
```

```python
data_pipe=FeatureUnion([
    ('obj_num',p1),
    ('obj_to_dum',p2),
    ('cyclic',p3),
    ('data',p4)
])
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
```

```python
params={'penalty':['l2'],'class_weight':['balanced',None],
        'C':np.linspace(0.01,100,100)}
```

```python
model=LogisticRegression()
```

```python
rs=RandomizedSearchCV(model,param_distributions=params,n_iter=10,
                      scoring='roc_auc',cv=10,n_jobs=-1,verbose=20)
```

### Splitting the data for train test and validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train0 = pd.concat([x_train,y_train], axis = 1)
```

```python
a_train, a_test = train_test_split(x_train0, test_size = 0.2, random_state=2)
```

```python
x_train1=a_train.drop(['Junk'],axis=1)
y_train1=a_train['Junk']
```

```python
x_train_test=a_test.drop(['Junk'],axis=1)
```

### Fitting the model on train data

```python
rs.fit(x_train1,y_train1)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.5f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
report(rs.cv_results_,5)
```

```python
lgrs=rs.best_estimator_
```

```python
lgrs.fit(x_train1,y_train1)
```
